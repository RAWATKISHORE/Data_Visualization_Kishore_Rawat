##Q3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.

##Ans.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate data
categories = ['A', 'B', 'C', 'D', 'E']
n_per_category = 50

data = []
for cat in categories:
    if cat == 'A':
        values = np.random.normal(100, 15, n_per_category)
    elif cat == 'B':
        values = np.random.normal(80, 10, n_per_category)
    elif cat == 'C':
        values = np.random.normal(120, 20, n_per_category)
    elif cat == 'D':
        values = np.random.normal(90, 12, n_per_category)
    else:  # cat == 'E'
        values = np.random.normal(110, 18, n_per_category)

    data.extend([(cat, val) for val in values])

# Create a DataFrame
df = pd.DataFrame(data, columns=['Category', 'Value'])

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Category', y='Value', ci='sd', capsize=0.1)
plt.title('Comparison of Categories (Bar Plot)')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Create a box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Category', y='Value')
plt.title('Comparison of Categories (Box Plot)')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Create a violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Category', y='Value')
plt.title('Comparison of Categories (Violin Plot)')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Print summary statistics
print(df.groupby('Category')['Value'].describe())
```

##Q4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

##Ans.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate data
categories = ['Low', 'Medium', 'High']
n_per_category = 300

data = []
for cat in categories:
    if cat == 'Low':
        values = np.random.normal(50, 15, n_per_category)
    elif cat == 'Medium':
        values = np.random.gamma(10, 5, n_per_category)
    else:  # cat == 'High'
        values = np.random.exponential(30, n_per_category) + 70

    data.extend([(cat, val) for val in values])

# Create a DataFrame
df = pd.DataFrame(data, columns=['Category', 'Value'])

# Create a box plot with violin plot overlay
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Category', y='Value', inner='box', cut=0)
plt.title('Distribution of Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Create a swarm plot with box plot overlay
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Category', y='Value', whis=[0, 100], width=.6, palette="vlag")
sns.swarmplot(data=df, x='Category', y='Value', color=".25", size=4, alpha=0.5)
plt.title('Distribution of Values Across Categories (with Individual Points)')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Create a faceted histogram
g = sns.FacetGrid(df, col="Category", height=5, aspect=.8)
g.map(sns.histplot, "Value", kde=True)
g.fig.suptitle('Distribution of Values by Category (Histograms)', y=1.05)
plt.show()

# Print summary statistics
print(df.groupby('Category')['Value'].describe())
```

##Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

##Ans.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate correlated data
n = 1000
feature_names = ['A', 'B', 'C', 'D', 'E']

# Create a correlation matrix
corr_matrix = np.array([
    [1.0, 0.8, -0.6, 0.1, 0.5],
    [0.8, 1.0, -0.5, 0.2, 0.4],
    [-0.6, -0.5, 1.0, -0.3, -0.2],
    [0.1, 0.2, -0.3, 1.0, 0.7],
    [0.5, 0.4, -0.2, 0.7, 1.0]
])

# Generate data from multivariate normal distribution
data = np.random.multivariate_normal(mean=[0, 0, 0, 0, 0], cov=corr_matrix, size=n)

# Create a DataFrame
df = pd.DataFrame(data, columns=feature_names)

# Compute the correlation matrix
corr = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()

# Create a pairplot for additional visualization
sns.pairplot(df, diag_kind='kde')
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()

# Print the correlation matrix
print("Correlation Matrix:")
print(corr)
```

##Q1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three- dimensional space.

np.random.seed(30)

data = {
'X': np.random.uniform (-10, 10, 300),
'Y': np.random.uniform (-10, 10, 300),
'Z': np.random. uniform (-10, 10, 300)

##Ans.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set random seed and create the dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create the 3D scatter plot
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
scatter = ax.scatter(df['X'], df['Y'], df['Z'], c=df['Z'], cmap='viridis', s=50, alpha=0.6)

# Customize the plot
ax.set_xlabel('X axis', fontsize=12, labelpad=10)
ax.set_ylabel('Y axis', fontsize=12, labelpad=10)
ax.set_zlabel('Z axis', fontsize=12, labelpad=10)
ax.set_title('3D Scatter Plot of Uniform Distribution', fontsize=16, pad=20)

# Add a color bar
cbar = fig.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Z value', rotation=270, labelpad=15, fontsize=10)

# Set the same limits for all axes
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)

# Add gridlines
ax.grid(True)

# Adjust the view angle
ax.view_init(elev=20, azim=45)

plt.tight_layout()
plt.show()

# Print summary statistics
print(df.describe())
```

##Q3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

np.random.seed(20)
data = {
'Month' np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
'Day' np.random.choice (range(1, 31), 100),
'Sales': np.random.randint(1000, 5000, 100) }
df = pd.DataFrame(data)


##Ans.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed and create the dataset
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix suitable for a heatmap
pivot_data = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap='YlOrRd', annot=True, fmt='.0f', cbar_kws={'label': 'Sales'})

# Customize the plot
plt.title('Sales Heatmap: Variation Across Months and Days', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Day', fontsize=12)

plt.tight_layout()
plt.show()

# Print summary statistics
print(df.groupby('Month')['Sales'].describe())
```

x = np.linspace (-5, 5, 100)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid
x, y = np.meshgrid(x, y)

# Calculate z = sin(√(x^2 + y^2))
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays and create a DataFrame
data = {'X': x.flatten(), 'Y': y.flatten(), 'Z': z.flatten()}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Set plot labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

##Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

counts = [20, 25, 30, 35]

##Ans.

```python
from bokeh.plotting import figure, show
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.io import output_file

# Data for the bar chart
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Specify the output file
output_file("fruit_counts_bar_chart.html")

# Create a bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', height=400, width=600)

# Add vertical bars
p.vbar(x='fruits', top='counts', width=0.6, source=source, legend_field='fruits',
       fill_color=factor_cmap('fruits', palette=['#FF5733', '#33FF57', '#3357FF', '#F0FF33'], factors=fruits))

# Customize the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

##Q4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist= np.random.randn(1000)

hist, edges= np.histogram (data_hist, bins=30)


##Ans.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(42)  # for reproducibility
data_hist = np.random.randn(1000)

# Calculate histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource for the histogram data
source = ColumnDataSource(data=dict(
    top=hist,
    left=edges[:-1],
    right=edges[1:]
))

# Create the figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           height=400, width=600)

# Add a quad glyph to represent the histogram bars
p.quad(top='top', bottom=0, left='left', right='right',
       fill_color="navy", line_color="white", alpha=0.5, source=source)

# Customize the plot
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Specify the output file
output_file("histogram.html")

# Show the plot
show(p)
```

##Q5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)

x = np.linspace (0, 1, 10)

y = np.linspace (0, 1, 10)

xx, yy = np.meshgrid(x, y)

##Ans.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColorBar, BasicTicker, LinearColorMapper
from bokeh.palettes import Viridis256

# Generate random data
np.random.seed(42)  # for reproducibility
data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Create the figure
p = figure(title="2D Heatmap",
           x_axis_label='X',
           y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           tools="pan,wheel_zoom,box_zoom,reset,save",
           width=600, height=500)

# Add the heatmap
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper,
                     ticker=BasicTicker(desired_num_ticks=10),
                     label_standoff=12,
                     border_line_color=None,
                     location=(0, 0))

p.add_layout(color_bar, 'right')

# Specify the output file
output_file("heatmap.html")

# Show the plot
show(p)
```
